#### Processando dados para comparação

```python
evolucao_total_confirmados = dados_plotar_evolucao_covid_19[ (dados_plotar_evolucao_covid_19['Tipo'] == 'Confirmed') ]
evolucao_total_mortos_recuperados = dados_plotar_evolucao_covid_19[ (dados_plotar_evolucao_covid_19['Tipo'] == 'Deaths') | (dados_plotar_evolucao_covid_19['Tipo'] == 'Recovered')]

evolucao_total_mortos_recuperados.reset_index(drop=True, inplace=True)
evolucao_total_confirmados.reset_index(drop=True, inplace=True)
```

#### Plotando gráfico comparativo entre as variáveis

```python
selection = alt.selection_multi(fields=['Date'])

escala1 = alt.selection_interval(bind='scales')
escala2 = alt.selection_interval(bind='scales')

top = alt.Chart(evolucao_total_mortos_recuperados, title="Comparativo entre recuperados e mortos").mark_bar().encode( 
    alt.Y('Date:O', scale=alt.Scale( domain=[ 'Jan 25 2020' , 'Feb 01 2020' , 'Feb 10 2020' , 'Feb 25 2020' , 'Mar 08 2020' , 'Mar 10 2020' , 'Mar 12 2020' ,'Mar 14 2020' ,'Mar 17 2020' , 'Mar 19 2020' ,'Mar 21 2020' ,'Mar 23 2020' ,'Mar 25 2020' ,'Mar 28 2020' ] ) ),
    x='Total:Q',
    color='Tipo:N',
).properties(
    width=600, height=200
).transform_filter(
    selection
).add_selection(
    escala2
)

bottom = alt.Chart(data=evolucao_total_confirmados, title="Casos confirmados no mundo entre 25 de Janeiro até 25 de Março").mark_bar().encode(
    alt.X('Date:O', scale=alt.Scale( domain=['Jan 25 2020' , 'Feb 01 2020' , 'Feb 10 2020' , 'Feb 25 2020' , 'Mar 08 2020' , 'Mar 10 2020' , 'Mar 12 2020' ,'Mar 14 2020' ,'Mar 17 2020' , 'Mar 19 2020' ,'Mar 21 2020' ,'Mar 23 2020' ,'Mar 25 2020' ,'Mar 28 2020' ] ) ),
    y='Total:Q',
    color=alt.condition(selection, alt.value('steelblue'), alt.value('lightgray'))
).properties(
    width=600, height=200
).add_selection(
    selection,
    escala1
)

alt.vconcat(top, bottom)
```

```python
class Timer:
    def __init__(self, timeout, callback):
        self._timeout = timeout
        self._callback = callback
        self._task = asyncio.ensure_future(self._job())

    async def _job(self):
        await asyncio.sleep(self._timeout)
        self._callback()

    def cancel(self):
        self._task.cancel()

def debounce(wait):
    def decorator(fn):
        timer = None
        def debounced(*args, **kwargs):
            nonlocal timer
            def call_it():
                fn(*args, **kwargs)
            if timer is not None:
                timer.cancel()
            timer = Timer(wait, call_it)
        return debounced
    return decorator
```
